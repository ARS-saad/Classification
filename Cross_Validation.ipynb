```python
import numpy as np
import xgboost as xgb
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split, KFold, StratifiedKFold, cross_val_score
import warnings
warnings.filterwarnings('ignore')
```

# Generating synthetic data & splitting into train-test

```python
x, y = make_classification(n_samples=1000, n_features=20, n_informative=2,
                           n_redundant=10, random_state=51)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=.3, random_state=52)
```

# XGBoost Classifier

```python
clf = xgb.XGBClassifier(use_label_encoder=False, eval_metric='logloss')
clf.fit(x_train, y_train)
```

```python
clf.score(x_test, y_test)
```

# Perform k-Fold Cross-Validation

```python
kf = KFold(n_splits=5, random_state=51, shuffle=True)
```

```python
kf_score = cross_val_score(clf, x, y, cv=kf)
kf_score
```

```python
kf_score.mean()
```

# Perform Stratified k-Fold Cross-Validation

```python
skf = StratifiedKFold(n_splits=5, random_state=60, shuffle=True)
skf_scores = cross_val_score(clf, x, y, cv=skf)
```

```python
skf_scores
```

```python
skf_scores.mean()
```
